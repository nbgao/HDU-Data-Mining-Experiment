### Z-Score标准化

```python
def ZScore(DT):
    m,n = np.shape(DT)
    DT_n = np.zeros((m,n))
    for i in range(m):
        u = np.mean(DT[i,:])
        s = np.std(DT[i,:])
        DT_n[i,:] = (DT[i,:]-u)/s

    return DT_n

DT1 = ZScore(DT)
```

```python
DT1
```

### 数据归一化

```python
def Standard(DT):
    m,n = np.shape(DT)
    DT_n = np.zeros((m,n))
    for i in range(m):
        Max = max(DT[i,:])
        Min = min(DT[i,:])
        DT_n[i,:] = (DT[i,:]-Min)/(Max-Min)
    return DT_n

DT2 = Standard(DT)
```

## 数据格式转换

```python
def DataTransfer(DT):
    n, m = np.shape(DT)
    data = []
    for i in range(m):
        key = range(n)
        value = DT[:,i]
        dd = dict(zip(key, value))
        data.append(dd)
    return data
```

```python
def LOO(data, label, model):
    Accuracy = []
    AccuracySum = 0
    for i in range(len(label)):
        p_label, p_acc, p_val = svm_predict(label[i:i+1], data[i:i+1], model);
        Accuracy.append(p_acc[0])
        AccuracySum += p_acc[0]
    AccuracyAverage = np.mean(Accuracy)
    return AccuracyAverage, Accuracy
```

### kernel = linear

```python
import copy
from sklearn.model_selection import KFold

count = 0
for i in range(m):
    feature_train = copy.copy(data)
    feature_train.remove(data[i])
    feature_test = data[i:i+1]
    
    label_train = copy.copy(label)
    label_train.remove(label[i])
    label_test = label[i:i+1]

    svm_model = svm_train(label_train, feature_train, '-c 4 -t 0')
    p_label, p_acc, p_val = svm_predict(label_test, feature_test, svm_model)
    if p_acc[0] == 100.0:
        count += 1
    
acc_average = count/m
print('\n正确率: %d/%d = %f' % (count, m, acc_average))
```
